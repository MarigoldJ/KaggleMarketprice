```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/new_sales_train.csv')
```

```python
by_cat = df.loc[:,['item_category_id','item_cnt_day']].sort_values(by = 'item_category_id', ascending = True)
```

```python
cat_sum = by_cat.groupby(by_cat['item_category_id']).sum().copy()
```

```python
cat_sum
```

```python
cat_sum.values.sum()
```

```python
cat_sum['item_cnt_day'] = cat_sum['item_cnt_day'] / 1617771
```

```python
cat_sum
```

```python
cat_sum['item_cnt_day'] = cat_sum['item_cnt_day'].round(4)
occ_dic = cat_sum.to_dict()
```

```python
cat_sum.reset_index(inplace=True)
cat_sum['item_cnt_day'] = cat_sum['item_cnt_day'] * 10000
cat_sum
```

```python
cat_sum.to_csv('./data/occ.csv', index=None)
```

```python
occ_dic = occ_dic['item_cnt_day']
occ_dic
```

```python
for i in occ_dic.keys():
    occ_dic[i] = occ_dic[i] * 10000
occ_dic
```

```python
for i in occ_dic.keys():
    occ_dic[i] = round(occ_dic[i],1)
occ_dic
```

```python
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['category_occ'] = np.zeros(shape=(len(df),), dtype=np.int32)
```

```python
for i in range(len(df)):
    df['category_occ'][i] = occ_dic[df['item_category_id'][i]]
```

```python
df.to_csv('./data/new_sales_train_occ.csv', sep=',')
```
