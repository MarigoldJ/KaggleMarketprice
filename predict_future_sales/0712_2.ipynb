```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/data_frame_v1.csv')
df
```

```python
g = df.groupby(['year', 'month', 'ID'], )
df.groupby(['year', 'month', 'ID']).sum()
```

```python
data = g.agg({
    'date_block_num': ['first'],
    'shop_id': ['first'],
    'item_id': ['first'],
    'item_category_id': ['first'],
    'item_price': ['mean'],
    'category_occ': ['first'],
    'mean_total': ['first'],
    'item_cnt_day': ['sum'],
})
data
```

```python
data.columns = ['data_block_num', 'shop_id', 'item_id', 'item_category_id', 'item_price', 'category_occ', 'total_mean_item_price', 'item_cnt_month']
data
```

```python
data.reset_index(inplace=True)
```

```python
data
```

```python
data.to_csv('./data/data_frame_v2.csv', index=None)
```

```python
data = pd.read_csv('./data/data_frame_v2.csv', index_col=None)
data
```

```python
x_train = data[data.data_block_num < 33].values[:, :10]
y_train = data[data.data_block_num < 33].values[:, 10]

x_test = data[data.data_block_num == 33].values[:, :10]
y_test = data[data.data_block_num == 33].values[:, 10]
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scale = scaler.transform(x_train)
x_test_scale = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
pd.datetime.now().strftime('%Y%m%d_%H%M')
```

```python
lr_model.fit(x_train_scale, y_train)
y_pred = lr_model.predict(x_test_scale)
print('Done')
```

```python
y_test.shape
```

```python
rmse = (((y_test - y_pred) ** 2).mean()) ** 0.5
rmse
```

```python
error = y_test - y_pred
print(error)
```
