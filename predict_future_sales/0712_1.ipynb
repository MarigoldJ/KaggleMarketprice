```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/new_sales_train_occ.csv', index_col=0)
```

```python
df.head()
```

```python
# datetime
df['date'] = pd.to_datetime(df['date'])

# 연도 추가
df['year'] = df['date'].dt.year
# 월 추가
df['month'] = df['date'].dt.month
# 공휴일 추가
import holidays
h = holidays.Russia()
def isholiday(date):
    weekend = date.weekday() in [5, 6]
    holiday = date in h
    if weekend or holiday:
        return 1
    else:
        return 0
df['holiday'] = df['date'].apply(isholiday)
df
```

```python
item_mean =  df.loc[:,['item_id', 'item_price']].groupby('item_id').mean()
```

```python
item_mean = item_mean.reset_index()
item_mean.columns = ['item_id','mean_total']
```

```python
item_mean.to_csv('./data/total_mean_item_price.csv', index=None)
```

```python
df = pd.merge(df, item_mean, how='left', on='item_id')
df.head()
```

```python
df.to_csv('data_frame_v1.csv',index=None)
```
